```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("AQ_01.txt", delimiter="\t", header=None)
#df.shape
df = df.iloc[:, :8]
df.columns = ["SP-12A", "SP-31", "TGS-813", "TGS-842", "SP-AQ3", "TGS-823", "ST-31", "TGS-800"]

# Plot
plt.figure(figsize=(14, 8))
for col in df.columns:
    plt.plot(df.index, df[col], label=col)

plt.title("Sensor Readings Over Time (AQ_01)")
plt.xlabel("Time (seconds)")
plt.ylabel("Resistance (kΩ)")
plt.legend()
plt.grid(True)
#plt.tight_layout()
plt.show()
```
